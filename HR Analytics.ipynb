# Percentage of No of Employees Doing Overtime

```python
Value=df['OverTime'].value_counts()
```

```python
plt.pie(Value,explode=[0.1,0.1],labels=['No','Yes'],autopct=lambda p:f'{p:.2f}%, {p*sum(Value)/100 :.0f}')
plt.title('Overtime Pie Chart')
```

```python
value=df['MaritalStatus'].value_counts()
```

```python
plt.pie(value,labels=['Married','Single','Divorced'],explode=[0.05,0.05,0.05],autopct=lambda p:f'{p:.2f}%, {p*sum(value)/100 :.0f}')
plt.title('Pie Chart For Marital Status')
```

```python
plt.figure(figsize=(10,5))
plot=sns.barplot(x=df['JobRole'],y=df.index)
plot.set_xticklabels(plot.get_xticklabels(), rotation=75)
plt.title('Count of Job Roles')
```
